```python
for i in lst:
    try:
        print("Current Element:", i)
        res = 5//i
        print("Result:" , res)
    except Exception as e:
        print('error occured', e)
        print(sys.exc_info()[0])
        
        print("=="*50)
        
print("\n 100 lines of code")
```

```python
for i in lst:
    try:
        print('Current Element : ', i)
        res = 5//int(i)
        print('result :', res)
    except ZeroDivisionError as ze:
        print("this is not a valid division:", ze)
    except ValueError as ve:
        print("Value error occured you can not divide interger by string :", ve)
    except Exception as e:
        print("Error happened:", e)
        print(sys.exc_info()[0])
    
    print("="*50)
    
    
```
