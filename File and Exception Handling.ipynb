```python
lst = [2,0,'hello', None]
for i in lst:
    try:
        print("current element:", i)
        res = 5/int(i)
        print("Result :",res)
    except Exception as e:
        print(e)
        print(sys.exc_info()[0])
    print("=="*50)
    
print("\n 100 lines of code")
```

```python
lst = [2,0,'hello', None]
```

```python
for i in lst:
    try:
        print("Current element :", i)
        res = 5/int(i)
        print("result:", res)
    except ZeroDivisionError as ze:
        print('its zero division error dont divide any number by zero')
    except ValueError as ve:
        print('you are making a value error')
    except Exception as e:
        print('error caused:', e)
        print(sys.exc_info()[0])
    print("=="*50)
    
print("\n 100 lines of code")
```
