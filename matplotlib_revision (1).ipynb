```python
import pandas as pd
```

```python

df = pd.read_csv("movie_meta.csv")
```

```python
df.columns
```

```python
len(df.movie_title[6].strip())
```

```python
def preprocess(x):
  return len(x.strip())
```

```python
df['movie_title_len'] = df.movie_title.apply(preprocess)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df.movie_title_len, bins=50)
plt.xticks(range(0,100,5))
plt.xlabel("Movie title len")
plt.ylabel("Number of movies - frequency")
plt.show()
```

```python
df
```

```python
df.movie_title_len
```

```python
d = {}
for i in df.movie_title_len:
  if i not in d:
    d[i] = 1
  else:
    d[i]+=1
```
